```python
from transformers import pipeline

# This will download the model to your local machine on first run
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-de-es")

text = "Ich möchte dieses Dokument ins Spanische übersetzen."
result = translator(text)

print(result[0]['translation_text'])
# Output: "Quiero traducir este documento al español."
```
